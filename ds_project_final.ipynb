```python
df['Price'] = df['Price'].str.replace('₹', '')
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].astype(np.int64)
df['Views'] = df['Views'].str.replace('(', '')
df['Views'] = df['Views'].str.replace(')', '')
df['Views'] = df['Views'].str.replace(',', '')
df['Views'] = df['Views'].astype(np.int64)
df.head()
```

```python
df['author/publication'] = df.apply(lambda row: row.info.split(",")[-1] , axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
import re
```

```python
def clean_description(text):
    return re.sub('[^a-zA-Z]',' ',text)
df['Description'] = df['Description'].apply(clean_description)
df.head()
```

```python
def lowercase(text):
    return str(text).lower()
df['Description'] = df['Description'].apply(lowercase)
df.head()
```

```python
import nltk
```

```python
def tokenization(text):
    return nltk.word_tokenize(text)
df['Description'] = df['Description'].apply(tokenization)
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words('english'))
def stopword(text):
    return [word for word in text if word not in stopwords]
df['Description'] = df['Description'].apply(stopword)
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
stemmer =WordNetLemmatizer()
def lemmetize(text):
    return [stemmer.lemmatize(word) for word in text]
df['Description'] = df['Description'].apply(lemmetize)
df.head()
```

```python
def cleaned_text(text):
    return " ".join(text)
df['Description']= df['Description'].apply(cleaned_text)
df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df['Description'])
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
new_df = pd.concat([df.Title,df.Rating,df.Views],axis=1)
# new_df=pd.DataFrame(df['Title'])
print(new_df.head())
new_df = new_df.reset_index()
del new_df['index']
new_df.info()
```

```python
index_of_title = pd.Series(new_df['Title'])
index_of_title
```

```python
cosine_sim.shape
```

```python
def recommend_books(title, cosine_sim = cosine_sim):
    recommended_books = []
    index = index_of_title[index_of_title == title].index[0]
    score = pd.Series(cosine_sim[index]).sort_values(ascending = False)
    top_5_book_index = list(score.iloc[1:6].index)
    
    for i in top_5_book_index:
        recommended_books.append(list(new_df['Title'])[i])
        
    return recommended_books
```

```python
recommend_books('The Rock Babas and Other Stories')
```

# Sample Results Below

```python
import random
for i in range(5):
    name =random.randint(0,373)
    print("Sample book name:",index_of_title[name])
    print("Top 5 recommended books:",recommend_books(index_of_title[name]))
    print('\n')
```
